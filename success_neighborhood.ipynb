```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("Analyzing Airbnb Geography...")
df = pd.read_csv('airbnb_ready_for_ml.csv', encoding='utf-8')

# ---------------------------------------------------------
# MANUAL ENCODING CORRECTION
# NOTE: These corrections were added retroactively. 
# After generating the plot for the first time, it was noticed that 
# the original data already had corrupted Portuguese characters (accents).
# Since it wasn't predictable which neighborhoods would make the Top 10 
# or how the errors would appear, we mapped only the visible cases 
# to fix the final chart presentation.
# ---------------------------------------------------------
neighborhood_corrections = {
    'Misericrdia': 'Misericórdia',
    'So Vicente': 'São Vicente',
    'Mina de gua': 'Mina de Água'
    # Add others as needed in the future
}

df['neighbourhood_cleansed'] = df['neighbourhood_cleansed'].replace(neighborhood_corrections)


# QUALITY FILTER
# NOTE: To ensure that we are not rewarding high turnover resulting from negative 
# experiences, listings with an average rating below 4.5.
# This guarantees that "Success" metric implies both volume and quality.

print("Filtering for high-quality listings (rating >= 4.5)...")
df = df[df['review_scores_rating'] >= 4.5]


# ANALYSIS AND CHART CREATION

# Group data by neighborhood and calculate average Success and Price
neighborhood_stats = df.groupby('neighbourhood_cleansed').agg(
    Total_Listings=('id', 'count'),
    Average_Success=('reviews_per_month', 'mean'),
    Average_Price=('price', 'mean')
).reset_index()

# Filter only neighborhoods with more than 50 listings
valid_neighborhoods = neighborhood_stats[neighborhood_stats['Total_Listings'] > 50]

# Select the top 10 neighborhoods with the highest Success
top_10_neighborhoods = valid_neighborhoods.sort_values(by='Average_Success', ascending=False).head(10)

# Create the Chart
plt.figure(figsize=(12, 6))
sns.barplot(data=top_10_neighborhoods, x='Average_Success', y='neighbourhood_cleansed', palette='magma')

plt.title('Top 10 Neighborhoods by Highest Turnover/Success', fontsize=14, fontweight='bold')
plt.xlabel('Average Monthly Reviews (Success)', fontsize=12)
plt.ylabel('Neighborhood', fontsize=12)
plt.tight_layout()
plt.savefig('03_success_per_neighborhood.png', dpi=300)
plt.close()

print("\n✅ Image '03_success_per_neighborhood.png' successfully generated with corrected names and quality filter applied!")
```
