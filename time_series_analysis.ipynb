```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("1. A carregar os dados temporais (Sazonalidade)...")
# Loading do clean dataset e dataset de reviews
df_listings = pd.read_csv('airbnb_ready_for_ml.csv')
df_reviews = pd.read_csv('reviews.csv')
```

```python
print("2. A cruzar as tabelas (Inner Join estilo SQL)...")
# O dataset de reviews usa 'listing_id'. O dataset clean usa 'id'.
# O 'inner join' garante que só irá analisar reviews de casas que ficaram após a limpeza!
df_merged = pd.merge(df_reviews, df_listings[['id']], left_on='listing_id', right_on='id', how='inner')
```

```python
print("3. A processar o formato de Tempo (Time Series)...")
# Converter a coluna de texto para formato Oficial de Data
df_merged['date'] = pd.to_datetime(df_merged['date'])
```

```python
# Extrair o Ano e o Mês
df_merged['Ano'] = df_merged['date'].dt.year
df_merged['Mes'] = df_merged['date'].dt.month

# Ver a partir de 2023 para observar a tendência atual do mercado
df_recentes = df_merged[df_merged['Ano'] >= 2023]
```

```python
# Agrupamento por Mês para descobrir a Curva Sazonal Anual
sazonalidade = df_recentes.groupby('Mes').size().reset_index(name='Total_Reservas')

print("4. A gerar o Gráfico de Séries Temporais...")
plt.figure(figsize=(12, 6))
```

```python
# Linha de tendência temporal
sns.lineplot(data=sazonalidade, x='Mes', y='Total_Reservas', marker='o', linewidth=3, color='#FF5A5F') # Cor oficial do Airbnb

# Formatar o Eixo X para mostrar os nomes dos meses em vez de números
meses_nomes = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
plt.xticks(range(1, 13), meses_nomes)

# Design do Gráfico
plt.title('Sazonalidade e Picos de Procura no Mercado (Proxy: Volume de Reviews)', fontsize=14, fontweight='bold')
plt.xlabel('Mês do Ano', fontsize=12)
plt.ylabel('Volume de Procura', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Encontrar o mês com o pico máximo automaticamente e desenhar uma linha
mes_pico = sazonalidade.loc[sazonalidade['Total_Reservas'].idxmax(), 'Mes']
plt.axvline(x=mes_pico, color='black', linestyle='--', alpha=0.5, label=f'Pico Máximo de Procura ({meses_nomes[mes_pico-1]})')
plt.legend()

plt.tight_layout()
plt.savefig('04_sazonalidade_series_temporais.png', dpi=300)
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Loading temporal data (Seasonality)

# Loading the clean dataset and reviews dataset
df_listings = pd.read_csv('airbnb_ready_for_ml.csv')
df_reviews = pd.read_csv('reviews.csv')
```

```python
# 2. Joining tables (SQL-style Inner Join)

# The reviews dataset uses 'listing_id'. The clean dataset uses 'id'.
# The 'inner join' ensures we only analyze reviews for listings that remained after cleaning!
df_merged = pd.merge(df_reviews, df_listings[['id']], left_on='listing_id', right_on='id', how='inner')
```

```python
# 3. Processing Time format (Time Series)

# Convert the text column to Official Date format
df_merged['date'] = pd.to_datetime(df_merged['date'])

# Extract the Year and Month
df_merged['Year'] = df_merged['date'].dt.year
df_merged['Month'] = df_merged['date'].dt.month

# Filter from 2023 onwards to observe the current market trend
df_recent = df_merged[df_merged['Year'] >= 2023]

# Group by Month to discover the Annual Seasonal Curve
seasonality = df_recent.groupby('Month').size().reset_index(name='Total_Bookings')
```

```python
# 4. Generating Time Series Chart

plt.figure(figsize=(12, 6))

# Temporal trend line
sns.lineplot(data=seasonality, x='Month', y='Total_Bookings', marker='o', linewidth=3, color='#FF5A5F') # Official Airbnb color

# Format the X-axis to show month names instead of numbers
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(range(1, 13), month_names)

# Chart Design
plt.title('Seasonality and Demand Peaks in the Market (Proxy: Review Volume)', fontsize=14, fontweight='bold')
plt.xlabel('Month of the Year', fontsize=12)
plt.ylabel('Demand Volume', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# Automatically find the month with the highest peak and draw a line
peak_month = seasonality.loc[seasonality['Total_Bookings'].idxmax(), 'Month']
plt.axvline(x=peak_month, color='black', linestyle='--', alpha=0.5, label=f'Highest Demand Peak ({month_names[peak_month-1]})')
plt.legend()

plt.tight_layout()
plt.savefig('04_seasonality_time_series.png', dpi=300)
plt.close()

print("\n✅ Chart '04_seasonality_time_series.png' generated successfully!")
```
