```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
```

```python
# 1. Loading the clean dataset

df = pd.read_csv('airbnb_ready_for_ml.csv')

# Final preparation: Remove pure text columns that ML cannot read (Neighborhoods and IDs)
# Convert 'room_type' (Private room or Entire home) into numerical variables (0 and 1)
df_ml = df.drop(columns=['id', 'neighbourhood_cleansed'])
df_ml = pd.get_dummies(df_ml, columns=['room_type'], drop_first=True)
```

```python
# CHART 1: CORRELATION MATRIX (What is linked to Success?)

print("2. Calculating the Correlation Matrix...")
plt.figure(figsize=(12, 10))

# Calculate the correlation of all variables
correlations = df_ml.corr()

# Focus only on the correlation with the success metric (reviews_per_month)
# Remove 'number_of_reviews' to avoid biasing the analysis
target_corr = correlations[['reviews_per_month']].drop(['reviews_per_month', 'number_of_reviews'])
target_corr = target_corr.sort_values(by='reviews_per_month', ascending=False)

sns.heatmap(target_corr, annot=True, cmap='coolwarm', fmt=".2f", vmin=-0.5, vmax=0.5)
plt.title('Correlation with Airbnb Success (Monthly Bookings)', fontsize=14, pad=15)
plt.tight_layout()
plt.savefig('01_correlation_matrix_airbnb.png', dpi=300)
plt.close()
```

```python
# CHART 2: MACHINE LEARNING (FEATURE IMPORTANCE)

# 3. Training the Artificial Intelligence (Random Forest)

# Define "X" (Listing features) and "y" (Success/Bookings)
# Remove review columns from X, as they are the outcome we want to predict
X = df_ml.drop(columns=['reviews_per_month', 'number_of_reviews'])
y = df_ml['reviews_per_month']

# Train the Model (100 decision trees thinking at the same time)
model = RandomForestRegressor(n_estimators=100, max_depth=10, random_state=42)
model.fit(X, y)

# Extract the importance of each feature (The algorithm's verdict)
importances = model.feature_importances_
df_importances = pd.DataFrame({
    'Feature': X.columns,
    'Decision_Weight': importances
}).sort_values(by='Decision_Weight', ascending=False)

# Draw the Model's Bar Chart

# 4. Generating the Feature Importance Chart

plt.figure(figsize=(12, 7))
sns.barplot(data=df_importances.head(10), x='Decision_Weight', y='Feature', hue='Feature', palette='viridis', legend=False)

plt.title('Feature Importance: What drives Airbnb Success? (Random Forest)', fontsize=14, fontweight='bold')
plt.xlabel('Algorithm Decision Weight (%)', fontsize=12)
plt.ylabel('Listing Features / Business Rules', fontsize=12)
plt.tight_layout()
plt.savefig('02_feature_importance_airbnb.png', dpi=300)
plt.close()

```
